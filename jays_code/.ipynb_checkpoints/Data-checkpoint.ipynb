```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt, mpld3
%matplotlib inline

filepath= '../data/ed_spending_per_student_10-15.csv'
data =  pd.read_csv(filepath)
data.head()
```

```python
new_data = data.drop(['Flag Codes'], axis=1)
```

```python
new_data.head(10)
```

```python
data_new = new_data.dropna()
data_new.head(10)
```

```python
data_new.to_csv('../clean_data/ED_doller_10-15.csv', index=False)
```

```python
df_time = data_new.groupby('TIME')
```

```python
df_time.first()
```

```python
 df_10 = df_time.get_group(2010)
 df_11 = df_time.get_group(2011)
 df_12 = df_time.get_group(2012)
 df_13 = df_time.get_group(2013) 
 df_14 = df_time.get_group(2014)
 df_15 = df_time.get_group(2015)
```

```python
df_10.head()
```

```python
# fig, ((ax1, ax2,ax3), (ax4,ax5, ax6)) = plt.subplots(3,3)
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(2,2)
ax1.bar(df_10, "LOCATION", "Value")
ax1.set(title = '2010 Education cost')
ax1.grid()


ax2.bar(df_11, "LOCATION", "Value")
ax2.set(title = '2011 Education cost')
ax2.grid()


ax3.bar(df_12, "LOCATION", "Value")
ax3.set(title = '2012 Education cost')
ax3.grid()


ax4.bar(df_10, "LOCATION", "Value")
ax4.set(title = '2013 Education cost')
ax4.grid()

plt.show()
```
