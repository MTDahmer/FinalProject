```python
%matplotlib inline
import pandas as pd

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import mpld3
mpld3.enable_notebook()


math= '../data/math_2010-2015.csv'
math_df =  pd.read_csv(math)

```

```python
Clean_math = math_df.drop(['Flag Codes'], axis=1)
```

```python
math_time = Clean_math.groupby(['TIME'])
math_country = Clean_math.groupby(['LOCATION'])
```

```python
df_12 = math_time.get_group(2012)

df_15 = math_time.get_group(2015)
```

```python
df15 = df_15.drop(18)
```

```python
n = 9
ind = np.arange(n)
label = ['Canada', 'Finland', 'France', 'Germany', 'Japan', 'Korea','Mexico','G Britain','USA']

width = 0.35  
fig = plt.figure()
plt.rcParams["figure.figsize"] = (8, 7)
bar1 = plt.bar(ind,df_12['Value'], width, label='2012')
bar2 = plt.bar(ind+width,df15['Value'], width, label='2015')

plt.ylabel('Math Scores', fontsize=14)
plt.title('USA Math Test Scores' , fontsize=14)

plt.xticks(ind + width / 2,label, fontsize=12, )


for rect in bar1 + bar2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom',size=11)

plt.legend(loc='best')
```

```python

mpld3.save_html(fig, 'math_bar.html', template_type="general")
```
