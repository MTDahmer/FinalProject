```python
%matplotlib inline
import pandas as pd

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import mpld3
mpld3.enable_notebook()



read =  pd.read_csv('../data/reading_2010-2015.csv')
read
```

```python
 clean_read= read.drop(['Flag Codes'], axis=1)
```

```python
read_time = clean_read.groupby(['TIME'])
read_country = clean_read.groupby(['LOCATION'])
```

```python
df_12 = read_time.get_group(2012)

df_15 = read_time.get_group(2015)
df_15
```

```python
n = 9
ind = np.arange(n)
label = ['Canada', 'Finland', 'France', 'Germany', 'Japan', 'Korea','Mexico','G Britain','USA']

width = 0.35  
fig = plt.figure()
plt.rcParams["figure.figsize"] = (9, 8)
bar1 = plt.bar(ind,df_12['Value'], width, label='2012', color='red')
bar2 = plt.bar(ind+width,df_15['Value'], width, label='2015', color='blue')

plt.ylabel('Reading Scores', fontsize=14)
plt.title('Reading Score Card' , fontsize=14)

plt.xticks(ind + width / 2,label, fontsize=12, )


for rect in bar1 + bar2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % int(height), ha='center', va='bottom',size=10)

plt.legend(loc='best')
```

```python
mpld3.save_html(fig, 'read_data.html', template_type="general")
```
